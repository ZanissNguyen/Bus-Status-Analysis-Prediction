```python
import pandas as pd
from datetime import datetime
from datetime import datetime

# Giả sử đây là Unix timestamp của bạn (đơn vị giây)
timestamp = 1742428811 

# Chuyển sang đối tượng datetime (mặc định là giờ địa phương)
dt_object = datetime.fromtimestamp(timestamp)

print("Ngày tháng:", dt_object)
# Output ví dụ: 2024-04-08 07:00:00
```

```python
import pandas as pd

# Create a dataframe with Unix timestamps from 12am to 12pm

# Generate Unix timestamps for every hour from 12am to 12pm (13 data points)
timestamps = [timestamp + (i * 3600) for i in range(13)]  # 3600 seconds = 1 hour

# Convert timestamps to datetime strings
datetimes = [datetime.fromtimestamp(ts).strftime('%d-%m-%Y %H:%M:%S') for ts in timestamps]

# Create dataframe with realtime column
df = pd.DataFrame({
    'realtime': datetimes,
    'unix_timestamp': timestamps
})

print(df)
```

```python
def convert_unix_to_realtime(df):
    """
    Converts the 'unix_timestamp' column in a dataframe to 'realtime' formatted datetime string.
    """
    df['realtime'] = df['unix_timestamp'].apply(lambda ts: datetime.fromtimestamp(ts).strftime('%d-%m-%Y %H:%M:%S'))
    return df
```

```python
def unix_to_datetime(df):
    """
    Converts a Unix timestamp (in seconds) to a formatted datetime string.
    """
    df['realtime'] = datetime.fromtimestamp(df['unix_timestamp']).dt.strftime('%d-%m-%Y %H:%M:%S')
    return df
```

```python
print(unix_to_datetime(df))
```
