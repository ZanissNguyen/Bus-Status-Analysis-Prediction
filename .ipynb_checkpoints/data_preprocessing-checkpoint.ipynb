```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import math
from tqdm import tqdm
import unicodedata
from datetime import datetime
import os
```

# LOAD BUS's STATUS DATA & STATION DATA

```python
bus_dataset = []

def get_waypoints(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)
        return [item["msgBusWayPoint"] for item in data]

bus_dataset.extend(get_waypoints("sample.json"))

folder = "./part2"
for filename in tqdm(os.listdir(folder)):
    if filename.endswith(".json"):
        bus_dataset.extend(get_waypoints(os.path.join(folder, filename)))

bus_df = pd.DataFrame(bus_dataset)
```

```python
print(len(bus_df))
bus_df.head()
```

```python
###Load station data
with open("bus_station.json", "r", encoding="utf-8") as f:
    station_dataset = json.load(f)

station_data = []
for route in station_dataset:
    stations = route['Stations']
    for station in stations:
        station['Lat'] = float(station['Lat'])
        station['Lng'] = float(station['Lng'])
        station_data.append(station)

station_df = pd.DataFrame(station_data)
print(station_df.head())
print(len(station_df))
```

# PROCESS BUS's STATUS DATAFRAME

```python
bus_df = bus_df.drop(["heading", "aircon", "working"], axis = 1)
bus_df.head(5)
```

```python
bus_df = bus_df[bus_df['door_up'].notna() | bus_df['door_down'].notna()]
len(bus_df)
```

```python
def unix_to_datetime(unix_time):
    return datetime.fromtimestamp(unix_time).strftime("%d-%m-%Y %H:%M:%S")

bus_df['realtime'] = bus_df['datetime'].apply(unix_to_datetime)

bus_df = bus_df.sort_values(['vehicle', 'datetime'])

bus_df = bus_df.drop_duplicates(subset=['vehicle', 'x', 'y'])

bus_df.head()
```

# PROCESS STATION DATAFRAME

```python
station_df.rename(columns={'Lat': 'y', 'Lng': 'x'}, inplace=True)
station_df.head()
```

# CALCULATE THE DISTANCE

```python
###Function that helps to calculate the distance between 2 point with lat and lng
def distance_calc(a, b):
    R = 6371
    
    y1 = math.radians(a['y'])
    x1 = math.radians(a['x'])
    y2 = math.radians(b['y'])
    x2 = math.radians(b['x'])
    
    dlat = y2 - y1
    dlng = x2 - x1
    
    a = math.sin(dlat/2)**2 + math.cos(y1)*math.cos(y2)*math.sin(dlng/2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    return R * c * 1000
```

```python
current_stations = []
station_distances = []
keep_mask = []

print("Length of bus's status data at first: ", len(bus_df))
for i in tqdm(range(len(bus_df))):
    status = bus_df.iloc[i]
    min_dis = 10000.0
    min_station = ""
    
    for j in range(len(station_df)):
        station = station_df.iloc[j]
        cur_dis = distance_calc(status, station)
        if cur_dis < min_dis:
            min_dis = cur_dis
            min_station = station['Name']
        if min_dis < 5:
            break
    if min_dis < 20:
        current_stations.append(str(min_station))
        station_distances.append(min_dis)
        keep_mask.append(True)
    else:
        keep_mask.append(False)

bus_df = bus_df[keep_mask].copy()
bus_df['current station'] = current_stations
bus_df['station distance'] = station_distances

print("Length of bus's status after being filtered and labeled: ", len(bus_df))
```

```python
bus_df.head()
```

# FILT BUS_DF BY DUPLICATE SAMPLES

```python
is_new_block = (bus_df['current station'] != bus_df['current station'].shift(1)) | \
               (bus_df['vehicle'] != bus_df['vehicle'].shift(1))

bus_df['block_id'] = is_new_block.cumsum()

idx_min_distance = bus_df.groupby('block_id')['station distance'].idxmin()

bus_df_cleaned = bus_df.loc[idx_min_distance].reset_index(drop=True)

bus_df_cleaned = bus_df_cleaned.drop(columns=['block_id'])

print(f"Data size at first: {len(bus_df)}")
print(f"Data size after being filted duplicate station data: {len(bus_df_cleaned)}")
bus_df_cleaned.head()
```

```python
bus_df_cleaned = bus_df_cleaned[bus_df_cleaned.groupby("vehicle")["vehicle"].transform("count") > 1]
bus_df_cleaned = bus_df_cleaned.reset_index(drop=True)
print(len(bus_df_cleaned))
bus_df_cleaned.head()
```

```python
def speed_calc(s, t):
    v = s / t * 3.6
    return v
```

```python
lst_data = []
prev_status = None
for i in tqdm(range(len(bus_df_cleaned))):
    cur_status = bus_df_cleaned.iloc[i]
    if prev_status is not None and prev_status['vehicle'] == cur_status['vehicle']:
        distance = distance_calc(prev_status, cur_status)
        d_time = cur_status['datetime'] - prev_status['datetime']
        speed = speed_calc(distance, d_time)
        lst_data.append({
            "start station": prev_status['current station'],
            "end station": cur_status['current station'],
            "start time": prev_status['realtime'],
            "end time": cur_status['realtime'],
            "distance (m)" : distance,
            "duration (s)": d_time,
            "speed (kmh)": speed 
        })
    prev_status = cur_status
```

```python
print(len(lst_data))
print(lst_data)
```

```python
main_df = pd.DataFrame(lst_data)
main_df.head()
```

```python
main_df.to_json("preprocessed_data.json", orient="records", force_ascii=False, indent=4)
```
